```python
import yfinance as yf
import datetime as dt
import matplotlib.pyplot as plt
import numpy as np

tickerSymbol = 'GC=F'
start_date = '2000-01-01' 
end_date = dt.datetime.now() 

tickerData = yf.Ticker(tickerSymbol)
name = tickerData.info['shortName']

tickerDf = tickerData.history(period='1d', start=start_date, end=end_date)

returns = tickerDf['Close'].pct_change().resample('Y').agg(lambda x: (x + 1).prod() - 1)

df_returns = returns.reset_index()

fig, ax = plt.subplots(figsize=(10, 6))

bar_colors = df_returns['Close'].apply(lambda x: 'g' if x >= 0 else 'r')

ax.bar(df_returns['Date'].dt.year, df_returns['Close'],
       color=bar_colors, edgecolor='black', linewidth=1)

ax.axhline(0, color='grey', linewidth=1, linestyle='--')

ax.set_title(f'Yearly Returns of {name}', fontsize=16, fontweight='bold')
ax.set_xlabel('Year', fontsize=12)
ax.set_ylabel('Returns', fontsize=12)

ax.set_xticks(np.arange(min(df_returns['Date'].dt.year), max(df_returns['Date'].dt.year) + 1, 1))
ax.set_xticklabels(df_returns['Date'].dt.year, rotation=45)

ax.yaxis.set_major_formatter(plt.FuncFormatter(lambda y, _: '{:.0%}'.format(y))) 

ax.yaxis.grid(True, linestyle='--', linewidth=0.5, alpha=0.7)

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.tight_layout()

plt.show()
```
