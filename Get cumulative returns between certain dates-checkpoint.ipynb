```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt


tickerSymbol = 'BTC-USD'
start_date = '2015-01-01'
end_date = '2017-12-15'

tickerData = yf.Ticker(tickerSymbol)
tickerDf = tickerData.history(period='1d', start=start_date, end=end_date)


daily_returns = tickerDf['Close'].pct_change().dropna()
cumulative_returns = (1 + daily_returns).cumprod() - 1

cumulative_returns.plot(figsize=(10, 6), title=f'Cumulative Returns of {tickerSymbol} from {start_date} to {end_date}')
plt.xlabel('Date')
plt.ylabel('Cumulative Returns')
plt.grid(True)
plt.show()

final_cumulative_return = cumulative_returns.iloc[-1]
print(f"The cumulative return of {tickerSymbol} from {start_date} to {end_date} was {final_cumulative_return:.2%}")
```
